```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

file_path = "Mall_Customers.csv"
data = pd.read_csv(file_path)

print(data.head())
```

```python

# Selecting features for clustering
features = data.iloc[:, [3, 4]].values  # Assuming "Annual Income" and "Spending Score"
k = 5
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(features)
centroids = kmeans.cluster_centers_

# Add the cluster labels to the original dataframe
data['Cluster'] = clusters
# Visualization
plt.figure(figsize=(8, 6))
sns.scatterplot(x=features[:, 0], y=features[:, 1], hue=clusters, palette="viridis", s=50, edgecolor='k')
plt.scatter(centroids[:, 0], centroids[:, 1], c='yellow', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering of Customers")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.show()
```
